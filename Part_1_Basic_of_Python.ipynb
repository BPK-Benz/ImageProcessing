**Example**) Print odds.

```python
question = range(15)
```

```python
odds = []
for i in question:
  if i % 2:
    odds.append(i)
odds
```

```python
odds = []
for i in question:
  odds += [i] if i % 2 else []
odds
```

```python
[i for i in question if i % 2]
```
